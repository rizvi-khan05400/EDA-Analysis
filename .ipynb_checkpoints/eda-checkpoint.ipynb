# Data PreProcessing

```python
# Replace white space into a "0" and converting "object" into a "float"
df["TotalCharges"] = df["TotalCharges"].replace(" ", "0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
# Convertin the value of "Senior Citizen"
def conv(value):
    if value == 1:
        return "yes"

    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

# Analysis

```python
# Creating count plot
plt.figure(figsize=(15, 6))
ax = sns.countplot(x = df["Churn"])
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn", fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Creating pie chart
plt.figure(figsize=(15, 6))
gb = df.groupby("Churn").agg({"Churn": "count"})
plt.pie(gb["Churn"], labels= gb.index, autopct= "%1.2f%%")
plt.title("Percentage of Churned Customeres", fontsize=16)
plt.show()
```

```python
# Checking "Churn" by "Gender"
plt.figure(figsize=(15, 6))
ax = sns.countplot(x="gender", data=df, hue="Churn")

# Add labels using bar containers
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', padding=3)

plt.title("Churn by Gender", fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Checking "Senior Citizen"
plt.figure(figsize = (15,6))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Prepare data
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Plot
fig, ax = plt.subplots(figsize=(15, 6))

# Plot the bars
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])

# Add percentage labels
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

plt.title("Churn by Senior Citizen (Stacked Bar Chart)", fontsize=16)
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage(%)")
plt.legend(title="Churn", bbox_to_anchor = (0.9,0.9))
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Making chart basis on "tenure"
plt.figure(figsize=(15, 6))
sns.histplot(x = "tenure", data = df, bins=72, hue = "Churn")
plt.title("Customers Who Stay With Us", fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Count on people basis on "Contract"
plt.figure(figsize=(15, 6))
ax = sns.countplot(x = "Contract", data = df, hue="Churn")

# Add labels using bar container
for container in ax.containers:
    ax.bar_label(container, fmt="%d", label_type="edge", padding=3)

plt.title("Count of Customers by Contract", fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Making subplot for multiple services
cols = ['PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies']

# Number of columns for the subplot grid
n_cols = 3
n_rows = (len(cols) + n_cols - 1) // n_cols # Calculate number of rows needed

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 5)) # Adjust figure size

# Flatten the axes array for easy iteration (handles both 1D and 2D arrays)
axes = axes.flatten()

# Iterate over columns and plot count plots
for i, col in enumerate(cols):
    ax = sns.countplot(x=col, data=df, ax=axes[i], hue=df["Churn"])

    # Add labels using bar container
    for container in ax.containers:
        ax.bar_label(container, fmt="%d", label_type="edge", padding=3)
    
    ax.set_title(f"Count Plot of {col}", fontsize=16)
    ax.set_xlabel(col, fontsize=14)
    ax.set_ylabel("Count", fontsize=14)
    ax.grid(axis='y', linestyle='--', alpha=0.7)

# Remove empty subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Checking "Churn" customers by "PaymentMethod"
plt.figure(figsize=(15, 6))
ax = sns.countplot(x = "PaymentMethod", data = df, hue="Churn")

# Add labels using bar container
for container in ax.containers:
    ax.bar_label(container, fmt="%d", label_type="edge", padding=3)

plt.title("Churned Customers by Payment Method", fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Conclusion:
Conclusion of Customer Churn EDA

1. Churn Rate

    A significant portion of customers have churned (~26–27%), which is concerning for business retention.

2. Demographics

    Gender has little to no impact on churn — churn is almost equally distributed across male and female customers.

    Senior Citizens show a much higher churn rate compared to younger customers.

3. Tenure (Customer Lifetime)

    Customers with shorter tenure (new customers) are far more likely to churn.

    Long-term customers (high tenure) show much lower churn.

4. Contract Type

    Month-to-month contracts have the highest churn.

    Customers with 1-year or 2-year contracts are far less likely to churn (they’re more “locked-in”).

5. Services Subscribed

    Customers lacking Online Security, Tech Support, or Backup services show higher churn.

    Internet service type also matters: Fiber-optic users have higher churn than DSL.

6. Payment Method

    Customers paying through electronic check churn the most.

    Automatic payments (credit card/bank transfer) reduce churn.

7. Business Insights

    Target senior citizens and month-to-month customers with retention offers.

    Encourage longer-term contracts through discounts or bundles.

    Promote value-added services (security, backup, tech support) to reduce churn risk.

    Shift customers from electronic checks to automatic payments for stability.
